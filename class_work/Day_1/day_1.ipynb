fillna():

    fillna() is a method that is used to fill NaN (null) values with a specified value or using a specified method.
    It can be applied to a specific DataFrame or Series.
    It can take various parameters such as the value to fill, method (e.g., 'ffill' for forward fill, 'bfill' for backward fill), and others.

fill():

    fill() is a more general method that can be used to fill NaN or missing values but is often used in the context of filling missing values by interpolation.
    It can be used for interpolation along a specified axis, and it has parameters like method and axis.

```python
price_pattern = r'Rs\.\s+([\d.]+)\s*-\s*([\d.]+)\s*Lakh'

# Extract and update 'Lower_Price' and 'Upper_Price' columns
df2[['Lower_Price', 'Upper_Price']] = df2['Price'].str.extract(price_pattern, expand=True)
df2['Lower_Price'] = pd.to_numeric(df2['Lower_Price'], errors='coerce')
df2['Upper_Price'] = pd.to_numeric(df2['Upper_Price'], errors='coerce')

# Display the updated DataFrame
print("Updated DataFrame:")
print(df2)
```
